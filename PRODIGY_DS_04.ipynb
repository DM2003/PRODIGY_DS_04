```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import zipfile
import io
```

```python
# Path to ZIP file
zip_file_path = "/content/TWITTER SENTIMENT ANALYSIS.zip"

# Name of the file we want to read
file_to_read = "twitter_training.csv"

# Opening the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Checking if the file exists in the ZIP archive
    if file_to_read in zip_ref.namelist():
        # Reading the file into memory
        file_data = zip_ref.read(file_to_read)
        # Converting bytes data to a StringIO object
        file_data_string = io.StringIO(file_data.decode('utf-8'))
        # Reading the StringIO object into a DataFrame
        twitter_training = pd.read_csv(file_data_string)
    else:
        print(f"File '{file_to_read}' not found in the ZIP archive.")
```

Now, we will use sentiment analysis tools  like Vader and TextBlob and understand the underlying patterns.

```python
# Using VADER
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Analyzing sentiment and appending compound scores to the dataframe
twitter_training['Compound'] = twitter_training['im getting on borderlands and i will murder you all ,'].apply(lambda x: sid.polarity_scores(x)['compound'])
brand_sentiment = twitter_training.groupby('Borderlands')['Compound'].mean().sort_values()

# Plotting sentiment according to brand
plt.figure(figsize = (10,8))
brand_sentiment.plot(kind = 'bar', color = 'skyblue')
plt.title("Average sentiment by brand")
plt.xlabel("Brand")
plt.ylabel("Average sentiment according to VADER score")
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

The higher the bar, the more positive the sentiment. The lower the bar, the more negative the sentiment.

For example, the brand "Borderlands" has an average sentiment score of 0.15, which is slightly positive. This means that, on average, tweets about Borderlands have a slightly positive sentiment.

The brand "The Division" has an average sentiment score of -0.04, which is slightly negative. This means that, on average, tweets about The Division have a slightly negative sentiment.


Overall, the sentiment is relatively neutral. There are slightly more positive tweets than negative tweets and the difference is being clearly seen.

```python
# Using TextBlob
from textblob import TextBlob
# We will use a function to get sentiment polarity using TextBlob
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Analyzing sentiment and appending polarity scores to the dataframe
twitter_training['Polarity'] = twitter_training['im getting on borderlands and i will murder you all ,'].apply(get_sentiment)

# Grouping by Brand and calculating average sentiment
brand_sentiment_textblob = twitter_training.groupby('Borderlands')['Polarity'].mean().sort_values()

# Plotting sentiment by Brand
plt.figure(figsize=(10, 8))
brand_sentiment_textblob.plot(kind='bar', color='lightgreen')
plt.title('Average Sentiment by Brand (TextBlob)')
plt.xlabel('Brand')
plt.ylabel('Average Sentiment (Polarity Score)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

These results suggest that there is a generally positive perception of "Borderlands" among Twitter users, while the sentiment towards "The Division" is slightly more negative. However, it's important to note that these are just average scores and there is still a significant amount of variation in sentiment within each brand.

Box plot

```python
 # Creating a box plot of sentiment scores for each brand
plt.figure(figsize=(12, 8))
sns.boxplot(x='Borderlands', y='Compound', data=twitter_training, palette='Set3')
plt.title('Distribution of Sentiment Scores by Brand')
plt.xlabel('Brand')
plt.ylabel('Sentiment Score (VADER Compound)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

 The box plot shows the distribution of sentiment scores for each brand.
     - The median sentiment score for "Borderlands" is higher than the median sentiment score for "The Division".
     - This suggests that there is a generally more positive sentiment towards "Borderlands" compared to "The Division".
     - The box plot shows that there is more variability in sentiment scores for "The Division" compared to "Borderlands".
     - This suggests that there is a wider range of opinions about "The Division" among Twitter users.

Overall, the box plot indicates that the sentiment towards "Borderlands" is generally positive, while the sentiment towards "The Division" is slightly more negative.
